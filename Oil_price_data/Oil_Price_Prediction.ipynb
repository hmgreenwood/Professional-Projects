```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()

    # Plotting rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    # Performing the Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
test_stationarity(oil_price_2010.Spot_Price_WTI)
```

We see that the Test Statistic is greater than the critical value, even at 10%, and therefore this data fails the stationarity test. We didn't really need to see this result at there is clearly a large decrease in price in 2015.

```python
multivar_model=price_impact.loc['01-01-2010':]
multivar_model.head()
```

```python
multivar_model.drop(['decade','year'],axis=1,inplace=True)
```

```python
#dropping 2020. The data is not yet complete and Covid is an oil "black swan event". Enhancing predictability in 2020 is not really useful
multivar_model=multivar_model.loc[:'2019-12-01']
multivar_model.tail()
```

```python
from statsmodels.tsa.vector_ar.vecm import coint_johansen
```

```python
coint_johansen(multivar_model,-1,1).eig
```

```python
multivar_model.tail()
```

```python
train=multivar_model.loc[:'2019-07-01']
test=multivar_model.loc['2019-08-01':]
```

```python
from statsmodels.tsa.vector_ar.var_model import VAR
```

```python
model = VAR(endog=train)
model_fit = model.fit()
```

```python
prediction = model_fit.forecast(model_fit.y, steps=len(test))
```

```python
from sklearn.metrics import mean_squared_error
cols=multivar_model.columns
pred = pd.DataFrame(prediction,index=test.index,columns=cols)
```

```python
for i in cols:
    
    print('RMSE of',i,'is:',np.sqrt(mean_squared_error(pred[i],test[i])))
```

```python
for i in cols:
    fig,ax=plt.subplots()
    plt.plot(pred[i],color='red',label='Model')
    plt.plot(test[i],label='Test',color='blue')
    plt.plot(train[i],label='Train',color='green')
    plt.title(i)
    plt.ylabel(i)
    plt.legend()
```

Assumptions and predictions made by VAR aren't out of line with what the expectation might have been. The past 5 months have obviously been quite volatile so accurate prediction is pretty difficult. The model predicted flat demand and production 

Using fbprophet with multiple variables

```python
X=multivar_model[['Total_Energy_Consumption','Mbbls_per_day','Mgal_per_day']]
y=multivar_model[['Spot_Price_WTI']]
```

```python
train_ds=pd.DataFrame()
train_ds=y
train_ds.reset_index(inplace=True)
train_ds.columns=['ds','y']
train_ds.head()
```

```python
X.reset_index(inplace=True)
```

```python
train_ds['Total_Energy_Consumption']=X['Total_Energy_Consumption']
train_ds['Mbbls_per_day']=X['Mbbls_per_day']
train_ds['Mgal_per_day']=X['Mgal_per_day']
```

```python
train_ds.tail()
```

```python
train_X=train_ds[:114]
test_X=train_ds[115:]
```

```python
prophet_reg=Prophet()
prophet_reg.add_regressor('Total_Energy_Consumption')
prophet_reg.add_regressor('Mbbls_per_day')
prophet_reg.add_regressor('Mgal_per_day')
#fitting the regressor
prophet_reg.fit(train_X)
future_data=prophet_reg.make_future_dataframe(periods=5)

forecast_data=prophet_reg.predict(test_X)
```

```python
prophet_reg.plot(forecast_data)
```

```python
print('rmse of tail end of 2019:', np.sqrt(mean_squared_error(forecast_data['yhat'],test_X['y'])))
```

```python
plt.figure(figsize=(12,8))
plt.plot(forecast_data['ds'],forecast_data['yhat'],label='Prophet Prediction',color='red')
plt.plot(test_X['ds'],test_X['y'],label='Actual Oil Price',color='green')
plt.legend()
```

The Prophet regression performs slightly worse than the VAR model, trained on the same data. To improve these models, additional training features should be added. Total US Production and gasoline sales do not tell all of the supply and demand story. Both models predict a falling oil price so maybe they were just ahead of their time.
